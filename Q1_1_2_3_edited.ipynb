```
import pandas as pd
import sqlite3

conn = sqlite3.connect('movies.db')
c = conn.cursor()

movies = pd.read_csv('movies_metadata.csv')
clean_movies = movies[~movies['budget'].str.contains('jpg')].copy()
clean_movies['budget'] = clean_movies['budget'].astype(int)
clean_movies.to_sql('clean_movies', conn, if_exists='replace', index = False)

ratings_small = pd.read_csv('ratings_small.csv')
ratings_small.to_sql('ratings_small', conn, if_exists='replace', index = False)
```

```
q1 = pd.read_sql_query(
"""
SELECT  
    title,
    budget
FROM clean_movies
WHERE budget > 5000000
""", 
    con= conn
)

q1
```

```
q2 = pd.read_sql_query("SELECT * FROM clean_movies WHERE original_language LIKE '%en%' AND title LIKE '% war %'", conn)
q2
```

```
q3 = pd.read_sql(
'''
SELECT 
   AVG(rating) AS 'Average Rating', 
   clean_movies.title As 'Title'
   FROM clean_movies
   LEFT JOIN ratings_small
        WHERE ratings_small.movieid = clean_movies.id
           GROUP BY clean_movies.title
''', con = conn)

q3
```
